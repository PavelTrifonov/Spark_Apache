```python
from pyspark import SparkContext

# Создание SparkContext
sc = SparkContext("local", "Longest_Sequence")

# Пример данных
data = [1, 2, 3, 4, 6, 7, 8, 10, 11, 12, 13, 14]

# Создание RDD из данных
rdd = sc.parallelize(data)

# Находим последовательности чисел и их длины
sequences = rdd.map(lambda x: (x, 1)) \
               .reduceByKey(lambda a, b: a + b) \
               .map(lambda x: (x[0], x[1], x[0] - x[1] + 1)) \
               .sortBy(lambda x: x[1], ascending=False)

# Получаем самую длинную последовательность
longest_sequence = sequences.first()

print("Самая длинная последовательность упорядоченных чисел:")
print("Начальное значение:", longest_sequence[2])
print("Конечное значение:", longest_sequence[0])
print("Длина последовательности:", longest_sequence[1])

# Завершение SparkContext
sc.stop()
```
